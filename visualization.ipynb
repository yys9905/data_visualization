```python
#read csv, csv파일 읽기
url="https://raw.githubusercontent.com/yys9905/data_visualization/main/2303.csv"
df=pd.read_csv(url,encoding='cp949')
```

```python
df["목록 등록일"] = pd.to_datetime(df['목록 등록일'])
```

```python
a = df['목록 등록일'].dt.year.value_counts().sort_index().reset_index(name="count")
```

```python
a
```

```python
sns.lineplot(data=a, x='index', y='count')
plt.title('연도별 데이터수')
plt.xlabel('년도')
plt.ylabel('데이터 수')
plt.show()
```
